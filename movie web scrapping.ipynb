```python
#In this project our aim is do to web scrapping of a movie website and fetch the required data into a csv file 
```

```python
#firstly, we will import all the required library
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd 
session = requests.session()
session.headers
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'}
```

```python
#printing all 50 pgs url into url_lst
url_lst=[]
base_url='https://www.themoviedb.org/movie?page='
for item in range(1,51):
    url_lst.append(base_url+str(item))
print(url_lst)
```

```python
common_url='https://www.themoviedb.org'
all_movie_list=[]
#for the below step we will put in all 50pgs url into a loop 
for url in url_lst:
    source_data=requests.get(url,headers=headers).text
    soup_data=BeautifulSoup(source_data,'lxml')
    all_divs=soup_data.find_all('div',class_='card style_1')
#now by taking all_divs we will find out each movies url 
    for i in all_divs:
        product_url=i.find('a')['href']
        product_url=common_url+product_url
#now by taking all movies url (ie product_url) we will further find our required data 
        movie_source=requests.get(product_url,headers=headers).text
        soup=BeautifulSoup(movie_source,'lxml')
        divs=soup.find_all('div',class_='header_poster_wrapper')
        for item in divs:
#now from divs we will find movie_name,rating,release_date,runtime,genre and director
            movie_name=item.find('a').text
#now we will use in try and except method, since at some places data is missing so by try if its present then the
#data will be printed or else in except it will take in null value if data is missing 
            try:
                rating=item.find('div',class_='user_score_chart')['data-percent']
            except:
                rating='null'
            try:
                release_date=item.find('span',class_='release').text.strip()[0:10]
            except:
                release_date='null'
            try:
                runtime=item.find('span',class_='runtime').text.strip()
            except:
                runtime='null'
            try:
                genre=item.find('span',class_='genres').text.strip().replace('\xa0','')
            except:
                genre='null'
            director_li=item.find_all('li',class_='profile')
            director='null'
            for i in range(0,len(director_li)):
                if 'Director' in director_li[i].find('p',class_='character').text:
                    director=director_li[i].find('a').text
    
                    
        
#now further, we will put in all our data into a dictionary                
        all_movie_info={
            'Name':movie_name,
            'Rating':rating,
            'Genre':genre,
            'Release date':release_date,
            'Runtime':runtime,
            'Director':director,
            'Url':product_url
        }
            
#next, we will append our dictionary(ie all_movie_info) into a list(ie all_movie_list) 
        all_movie_list.append(all_movie_info)
                
                
                
            
    
                
        
    

        
    

       


```

```python
#now that we have got in all our data we will print it 
print(all_movie_list)
```

```python
#using pandas will we will put in our all_movie_list into a varible(df)
df=pd.DataFrame(all_movie_list)
```

```python
#we will put in all our data in csv file and name it as allmovie
df.to_csv('allmovie.csv')
```
